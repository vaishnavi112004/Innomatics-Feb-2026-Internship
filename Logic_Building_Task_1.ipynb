```python
# 1. User Login Check


username = "admin"
password = "1234"

# Take user input
entered_user = input("Enter username: ")
entered_pass = input("Enter password: ")

# Check if both username and password match
if entered_user == username and entered_pass == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# 2. Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Count pass and fail students
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Display final result
print("Total Passed Students:", pass_count)
print("Total Failed Students:", fail_count)
```

```python
# 3. Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Remove spaces and convert to lowercase
for name in names:
    cleaned_names.append(name.strip().lower())

print("Cleaned Names List:", cleaned_names)
```

```python
# 4. Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message
for msg in messages:
    length = len(msg)
    print("Message:", msg)
    print("Length:", length)

    # Flag messages longer than 10 characters
    if length > 10:
        print("Flag: Long message")

    print()
```

```python
# 5. Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Count number of "ERROR" entries
for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
