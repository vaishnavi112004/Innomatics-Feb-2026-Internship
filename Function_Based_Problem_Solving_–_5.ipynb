```python
# Problem 1: Social Media – Post Engagement Analyzer
def analyze_engagement(likes_list):
    total_likes = 0
    for likes in likes_list:
        total_likes += likes
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    print("Total Likes:", total_likes)
    print("Post Status:", status)

analyze_engagement([200, 300, 250, 300])
```

```python
# Problem 2: Healthcare – Medicine Stock Alert System
def check_medicine_stock(stock):
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"
    print("Medicine Stock:", stock)
    print("Status:", status)

check_medicine_stock(6)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
def rainfall_checker(rainfall_list, required_level):
    total = 0
    for rain in rainfall_list:
        total += rain
    average = total / len(rainfall_list)
    status = "Adequate Rainfall" if average >= required_level else "Inadequate Rainfall"
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rainfall_checker([60, 70, 80, 78], 70)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
def detect_duplicates(usernames):
    unique_users = set(usernames)
    status = "Yes" if len(unique_users) != len(usernames) else "No"
    print("Duplicate Accounts Found:", status)

detect_duplicates(["user1", "user2", "user1"])
```

```python
# Problem 5: Healthcare – Appointment Eligibility Checker
def appointment_eligibility(age):
    status = "Eligible" if age >= 18 else "Not Eligible"
    print("Patient Age:", age)
    print("Eligibility Status:", status)

appointment_eligibility(21)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
def premium_crops(prices):
    premium = [price for price in prices if price > 2000]
    print("Premium Crops:", premium)

premium_crops([1500, 2500, 1800, 3200])
```

```python
# Problem 7: System Monitoring – Application Health Checker
def system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print("Error Count:", errors)
    print("System Status:", status)

system_health(7)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker
def transaction_check(amount):
    limit = 50000
    status = "Approved" if amount <= limit else "Rejected"
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_check(60000)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System
def attendance_eligibility(attendance_list):
    present = 0
    for day in attendance_list:
        if day == "P":
            present += 1
    percentage = (present / len(attendance_list)) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_eligibility(["P", "P", "A", "P", "P"])
```

```python
# Problem 10: Smart Electricity Bill Analyzer
def electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Classification:", usage)

electricity_bill(250)
```
