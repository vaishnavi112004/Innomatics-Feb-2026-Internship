```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount: â‚¹"))

            # Check minimum amount
            if amount < 50:
                print("âŒ Amount must be at least â‚¹50. Try again.")
                continue

            # Check valid plan
            if amount in valid_plans:
                print("âœ… Recharge Successful!")
                break
            else:
                print("âŒ Invalid plan. Available plans:", valid_plans)

        except ValueError:
            print("âŒ Please enter a valid number.")

# Call function
validate_recharge()
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(f"{product}: ðŸ”´ Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}: ðŸŸ¢ Stock OK (Stock = {quantity})")

# Sample dictionary
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Wheat": 18
}

# Call function
check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example usage
student_marks = [60, 70, 45, 80, 55]
result = process_result(student_marks)
print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator

def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    total_fare = base_fare + (distance * per_km)

    if peak.lower() == "yes":
        total_fare += total_fare * 0.25  # 25% extra

    return total_fare


while True:
    try:
        km = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ")

        fare = calculate_fare(km, peak_hour)
        print("Total Fare: â‚¹", fare)

        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            break

    except ValueError:
        print("âŒ Please enter valid input.")
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    present_days = 0

    # Count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example usage
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
status = check_attendance(attendance)
print("Status:", status)
```

```python
# Problem 6: Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Check length
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    # Loop through password
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include digit & special character)"

# Example usage
pwd = input("Enter password: ")
print(check_password(pwd))
```
