### Python Programming Assignment – Data Processing and Analysis




```python
# Problem Statement 1: Employee Performance Bonus Eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Get all employees who have the highest score (handles ties)
top_performers = [name for name, score in employees.items() if score == highest_score]

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
# Problem Statement 2: Search Query Keyword Analysis

import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase and remove punctuation
query = query.lower()
query = query.translate(str.maketrans('', '', string.punctuation))

words = query.split()
frequency = {}

# Count frequency of each keyword
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Keep only keywords searched more than once
result = {word: count for word, count in frequency.items() if count > 1}

print(result)
```

```python
# Problem Statement 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Store even readings as (hour_index, value)
for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((hour, value))

print("Valid Sensor Readings (Hour, Value):", valid_readings)
```

```python
# Problem Statement 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Extract domain from each email and count occurrences
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_emails = len(emails)

# Calculate and print percentage usage
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem Statement 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate the average daily sales
average_sales = sum(sales) / len(sales)

# Define threshold as 30% above the average
threshold = average_sales * 1.30

# Check each day's sales against the threshold
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")

```

```python
# Problem Statement 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Count how many times each user ID appears
for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

# Display only duplicate IDs
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
